# 07. cleaning up orders dataframe

testing for any mixed data types

```python
for col in df_ords.columns.tolist():
  weird = (df_ords[[col]].map(type) != df_ords[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_ords[weird]) > 0:
    print (col)
```

checking for any null cells

```python
df_ords.isnull().sum()
```

no missing or null values - the high number of null values in the last column is explained by the fact that a customer's first order will receive a null value. 

```python
df_ords_nan = df_ords[df_ords['days_since_prior_order'].isnull() == True]
```

```python
df_ords_nan
```

changing the nan (or missing values) to 0

```python
df_ords['days_since_prior_order'] = df_ords['days_since_prior_order'].fillna(0)
df_ords
```

adding in an extra column to explain the 0 value in the column days since prior order - is it the first order yes/no

```python
df_ords.loc[df_ords['days_since_prior_order'] == 0, "first_order"] = "yes"
df_ords.loc[df_ords['days_since_prior_order'] >= 1, "first_order"] = "no"
df_ords
```

finding duplicate values

```python
df_dups = df_ords[df_ords.duplicated()]
```
