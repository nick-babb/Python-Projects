Aggregating product prices using transform function

```python
df_ords_prods_merge['avg_product_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

Checking the result

```python
df_ords_prods_merge[['user_id', 'avg_product_price']].head(60)
```

Creating a spend flag for average product price

```python
def spend_flag(row):

  if row['avg_product_price'] < 10:
    return 'Low spender'
  elif row['avg_product_price'] >= 10:
    return 'High spender'
  else: return 'Not enough data'
```

Applying the flag to the df

```python
df_ords_prods_merge['spend_flag'] = df_ords_prods_merge.apply(spend_flag, axis=1)
```

```python
df_ords_prods_merge[['user_id', 'avg_product_price', 'spend_flag']].head()
```

Checking the frenquency of the customer spend category variable

```python
df_ords_prods_merge['spend_flag'].value_counts(dropna = False)
```

Creating an additional variable for the median days since the last order

```python
df_ords_prods_merge['median_days_since_prior_order'] = df_ords_prods_merge.groupby('user_id')['days_since_prior_order'].transform(np.median)
```

Creating a frequency flag

```python
def frequency_flag(row):

  if row['median_days_since_prior_order'] > 20:
    return 'Non-frequent customer'
  elif row['median_days_since_prior_order'] <= 10:
    return 'Frequent customer'
  elif (row['median_days_since_prior_order'] > 10) and (row['median_days_since_prior_order'] <= 20) :
    return 'Regular customer'
  else: return 'Not enough data'
```
